```python
fig, ax = plt.subplots(figsize=(15, 7))
for n, col in enumerate(df["Category"].unique()):
    ax.boxplot(df[df["Category"]==col]["Calories"], positions=[n+1])

x_labels = ["0"]
x_labels.extend(df["Category"].unique())

ax.set_xticks(range(10))
ax.set_xticklabels(x_labels)
ax.set_title("Box plot of calories per category")
plt.show()
```

we can observe the outlier under "Chicken & Fish" category again in this boxplot as well, and we observe that the category with the largest variance in calories is "Smoothies & Shakes", this is probably because this category contains items with different sizes (Small/medium/large) which will increase the overall variance of calories within the cateogry


## Lets explore the outlier

```python
df[df["Category"] == "Chicken & Fish"].sort_values(by="Calories", ascending=False).head()

f, ax = plt.subplots(figsize=(18,5))

plt.barh(
    y=df[df["Category"] == "Chicken & Fish"].sort_values(by="Calories", ascending=False).head()["Item"],
    width=df[df["Category"] == "Chicken & Fish"].sort_values(by="Calories", ascending=False).head()["Calories"],
)
```

The outlier in the "Chicken & Fish" category seems to be a sharing meal "(40 piece) of Chicken Nuggets"

```python
df["Serving Size (metric)"] = df["Serving Size"].apply(lambda x: x.split("(")[-1][:-1][:-1].split()[0] if "(" in x else "")
```

There are some "Serving sizes" that do not contain the metric unit equivilant

```python
df[~df["Serving Size"].str.contains('\(')]["Serving Size"].unique()
```

Since we cannot convert directly from "fl oz" to "g" without knowing the density of the item, we will ignore the items that didn't include the service size in grams for the next visualization

```python
df["Serving Size (metric)"] = pd.to_numeric(df["Serving Size (metric)"])
```

```python
fig, ax = plt.subplots(figsize=(15, 7))

colors = {cat:color for cat, color in zip(
    df["Category"].unique(), 
    ["blue", "orange", "green", "red", "purple", "brown", "pink", "gray", "olive"])
}

for cat, col in colors.items():
    ax.scatter(
    x=df[df["Category"]==cat]["Serving Size (metric)"], 
    y=df[df["Category"]==cat]["Calories"],
    c=df[df["Category"]==cat]["Category"].map(colors),
    edgecolors='black',
    s=100,
    label = cat
    )
ax.legend()
ax.set_xlabel("Serving size")
ax.set_ylabel("Calories")
ax.set_title("Scatter plot of Calories vs Serving Size")
```

The general trend shows that there is positive correlation between serving size and calories (as expected).

and Salads as well as Chicken & Fish generally has less calories per serving size compared to the overall average trend


### Are there any items that exceeds the daily % value?

```python
([x for x in df.columns if "%" in x])
```

```python
f, ax = plt.subplots(figsize=(18,5))

labels = ['Total Fat (% Daily Value)',
 'Saturated Fat (% Daily Value)',
 'Cholesterol (% Daily Value)',
 'Sodium (% Daily Value)',
 'Carbohydrates (% Daily Value)',
 'Dietary Fiber (% Daily Value)',
 'Vitamin A (% Daily Value)',
 'Vitamin C (% Daily Value)',
 'Calcium (% Daily Value)',
 'Iron (% Daily Value)']

ax.bar(
    x= labels,
height=df[['Total Fat (% Daily Value)',
 'Saturated Fat (% Daily Value)',
 'Cholesterol (% Daily Value)',
 'Sodium (% Daily Value)',
 'Carbohydrates (% Daily Value)',
 'Dietary Fiber (% Daily Value)',
 'Vitamin A (% Daily Value)',
 'Vitamin C (% Daily Value)',
 'Calcium (% Daily Value)',
 'Iron (% Daily Value)']].agg("max").values
    )
# ax.tick_params(axis='x', rotation=90)

plt.xticks(range(10), [textwrap.fill(label, 10) for label in labels], 
fontsize=12, horizontalalignment="center")

thresh = 100
ax.axhline(100, color="red")
ax.text(1.02, 100, "100%\nthreshold", va='center', ha="left", bbox=dict(facecolor="w",alpha=0.5),
        transform=ax.get_yaxis_transform())
ax.set_title("Bar chart of % Daily value for various nutrients")
```

We observe that there are some items that exceed the daily% value for the following:
- Total Fat
- Cholesterol
- Sodium
- Vitamin A
- Vitamin C

lets figure out these items:

```python
df[
    (df["Total Fat (% Daily Value)"]>100) | 
    (df["Cholesterol (% Daily Value)"]>100) |
    (df["Sodium (% Daily Value)"]>100) |
    (df["Vitamin A (% Daily Value)"]>100) |
    (df["Vitamin C (% Daily Value)"]>100)
    ].sort_values(by="Calories", ascending=False)
```

There are 16 items that fits the criteria... knowing that the median Calories = 340, lets focus on
the food items that has more than 340 calories 

```python
df[
    ((df["Total Fat (% Daily Value)"]>100) | 
    (df["Cholesterol (% Daily Value)"]>100) |
    (df["Sodium (% Daily Value)"]>100) |
    (df["Vitamin A (% Daily Value)"]>100) |
    (df["Vitamin C (% Daily Value)"]>100)) & (df["Calories"]>340)
    ].sort_values(by="Calories", ascending=False)
```

These 6 items are the worst items in the menu because they are high in calories and they exceed the daily % value in at least 1 category of nutrients.
